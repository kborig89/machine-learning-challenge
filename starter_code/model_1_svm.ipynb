```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]


# Set features. This will also be used as your x values.
selected_features = df[['koi_disposition','koi_period','koi_period_err1','koi_period_err2',
                        'koi_time0bk','koi_time0bk_err1','koi_time0bk_err2',
                        'koi_impact','koi_impact_err1','koi_impact_err2',
                        'koi_duration','koi_duration_err1','koi_duration_err2',
                        'koi_depth','koi_depth_err1','koi_depth_err2',
                        'koi_prad','koi_prad_err1','koi_prad_err2',
                        'koi_teq','koi_insol','koi_insol_err1','koi_insol_err2',
                        'koi_model_snr','koi_steff','koi_steff_err1','koi_steff_err2',
                        'koi_slogg','koi_slogg_err1','koi_slogg_err2',
                        'koi_srad','koi_srad_err1','koi_srad_err2',
                        'ra','dec','koi_kepmag']]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
 from sklearn.model_selection import train_test_split
X = selected_features.drop("koi_disposition")
y = selected_features["koi_disposition"]
print(X.shape, y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_minmax = MinMaxScaler().fit(X_train)

X_train_minmax_scaled = X_minmax.transform(X_train)
X_test_minmax_scaled = X_minmax.transform(X_test)
```

# Train the Model



```python
 # Support vector machine linear classifier
from sklearn.svm import SVC 
model2 = SVC(kernel='linear')
model2.fit(X_train_minmax_scaled, y_train)

#from sklearn.svm import LinearSVC
#clf = LinearSVC(random_state=0, tol=1e-5)
#clf.fit(X_train, y_train.ravel()) 

print(f"Training Data Score: {model2.score(X_train_minmax_scaled, y_train)}")
print(f"Testing Data Score: {model2.score(X_test_minmax_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.001, 0.01]}
grid2 = GridSearchCV(model2, param_grid, verbose=3, n_jobs=-1)
```

```python
# Train the model with GridSearch
grid2.fit(X_train, y_train)
```

```python
print(grid2.best_params_)
print(grid2.best_score_)
```
